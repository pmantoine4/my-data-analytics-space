```python
#To drop the '?' values in the race column and to verify the effectiveness of this cleaning operation
df_new=df.drop(df[df['race']=='?'].index)
df_new.head()
```

```python
#To filter the dataset and create a subset of seven columns necessary for the analysis
df2 = df_new[['encounter_id','patient_nbr','race','gender','age','time_in_hospital','diabetesMed']]
df2.head()
```

```python
#To calculate, applying the groupby function, the number of hospitalizations per race
df2.groupby(['race'])['encounter_id'].count().sort_values(ascending=False)
```

```python
#To visualize the result of the above operation
color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.6 # transparency
         )

df2.groupby(['race'])['encounter_id'].count().sort_values(ascending=False).plot(kind='bar',color=color)
plt.title('Comparison of hospitalizations number due to diabetes with regard to race')
```

```python
#A pie chart for the same above result
df2.groupby(['race'])['encounter_id'].count().sort_values(ascending=False).plot(kind='pie')
plt.title('Comparison of hospitalizations number due to diabetes with regard to race')
```

```python
#To calculate, applying the groupby function, the number of hospitalizations per gender
df2.groupby(['gender'])['encounter_id'].count().sort_values(ascending=False)
```

**Conclusion**
The original dataset contains 101766 rows and 50 columns. After a cleaning process, the number of rows was reduced to 99493. Additionally, for the purposes of analysis, a subset was created with the following columns: encounter_id, patient_nbr, race, gender, age, time_in_hospital, and diabetesMed .
The analysis of the data made it possible to arrive at the following conclusions:
-Patients of so-called Caucasian race are by far the most numerous to be hospitalized due to diabetes, 76,099 cases or 76.48%. In second place come African Americans with 19,210 cases or 19.31%. Asians are the least present with 641 cases or 0.64%.
-In relation to gender, female hospitalizations, 53,575 or 53.84%, exceed those of men who show 45,917 cases or 46.15%.
-In relation to age, it is the 70-80 year olds who come out on top with 25,469 cases or 25.60%. 0-10 year olds are the least numerous, 160 cases or 0.16%.
-A second measure was analyzed: the time spent in the hospital by patients. In this case, it is African Americans who come first with an average stay of 4.5 days per hospitalization. Last place is occupied by Asians with an average of 3.99 days. On the other hand, women with an average stay of 4.47 days exceed men who have an average of 4.30 days. In terms of age, it is those aged 80-90 who dominate the ranking with an average stay of 4.80 days.
-The last parameter to be analyzed is whether or not antidiabetic medications were taken during hospitalization. Caucasians come out on top both for taking medication, 58,450 cases, and for not taking it, 17,649 cases. In relation to gender, women dominate both for taking, 40,867 cases or 41.07%, and for not taking medications, 12,708 cases or 12.77%. From the point of view of age, 70-80 year olds occupy the first class of the ranking with 19,675 cases or 19.77%, followed by 60-70 year olds with 17,212 cases or 17.30%.
