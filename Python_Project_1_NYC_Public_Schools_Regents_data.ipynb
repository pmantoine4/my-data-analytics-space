```python
#To create a subset of the dataset for the requested task (analysis). We will name it df2. (filtering columns)
df2 = df[['School DBN','School Name','School Level','Regents Exam','Year','Total Tested','Mean Score']]
```

```python
#To view and understand df2
df2.head()
```

```python
#Processing the data in the Mean Score column by replacing the string 's' value by its numeric equivalent '75' (Rf. NYC, DOE: High School Academic Policy Guide)
df2["Mean Score"]=df2["Mean Score"].replace("s","75")
NB.
When we filter rows in the dataset by applying the code  df2[df2['Mean Score']=='s'] , we obtain a count of 75222 rows (35.42% of total rows in the dataset), which is very important. It's a choice of the dataset owner to have 's' as values in the dataset (Rf. dataset dictionary). But, instead of dropping all these values that could affect the relevance  of the analysis,  we have opted, regarding the "Mean Score" column, to inpute "75" as the numeric value of "s" as indicated in the E-U scale, in accordance with  the " NYC, DOE: High School Academic Policy Guide"
```

```python
#To verify that 's' has been effectively replaced by '75', (but the data type is still object that means string)
df2['Mean Score']
```
