```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#To download the dataset
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/Salaries.csv')
```

```python
#To understand the dataset
df.head()
```

```python
#To obtain the summary stats of the dataset
df.describe()
```

```python
#To obtain the median of the dataset
df.median()
```

```python
#To obtain the number of rows and columns in the dataset
df.shape
```

```python
#Grouping the data by rank, and finding the median salary for each rank/title
df.groupby(['rank'])['salary'].median()
```

```python
#Grouping the data by sex, and finding the median salary for each genre
df.groupby(['sex'])['salary'].median()
```

```python
#To clean the dataset by naming the missing column name
df=df.rename(columns={'Unnamed: 0':'Unique Key'})
```

```python
#To confirm  the dataset has been cleaned by naming the missing column name
df.head()
```

```python
#To create a new dataframe that consists of a subset of the original columns (Unique Key, rank, yrs.service, sex, salary)
df2=df[['Unique Key','rank','yrs.service','sex','salary']]
```

```python
#To confirm that the subset has been created
df2.head()
```

```python
#To verify the existence or not of missing data
df2.count()
```

```python
#Grouping the data by rank in the subset (df2) and couting it based on Unique Key
df2.groupby(['rank'])['Unique Key'].count().sort_values(ascending=False)
```

```python
#Ploting the result obtained above
df2.groupby(['rank'])['Unique Key'].count().sort_values(ascending=False).plot(kind='bar')
```

```python
#Grouping the data by sex in the subset (df2) and couting it based on Unique Key
df2.groupby(['sex'])['Unique Key'].count().sort_values(ascending=False)
```

```python
#Ploting the result obtained above
df2.groupby(['sex'])['Unique Key'].count().sort_values(ascending=False).plot(kind='bar')
```

**CONCLUSIONS**

    -When we group the data  by genre (sex), the median salary is 103,750.00 for the female profs and 108,043.00 for the male ones. These figures show an inegality of treatment regarding the salary related to the genre. The data shows also that the presence of male profs is very dominant inside the institution (358 or 90.18%) while the female profs account for only
    9.82% of the total (39).
